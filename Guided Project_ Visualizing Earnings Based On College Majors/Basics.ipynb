### Analysing the plots

1. Do students in more popular majors make more money?

The popularity of a course is measured by its sample size. Then, by looking to the 'Sample Size' x 'Median Salary' plot, it's possible to conclude that a higher popularity is not correlated with higher salaries.

2. Do students that majored in subjects that were majority female make more money?

The 'ShareWomen' value measures the percentage of female students that graduated from that major. Then, by looking to the 'ShareWomen' x 'Median Salary' plot, it's possible to notice that students that majored in subjects that were majority female, i.e with ShareWomen > 0.5, do not make more money.

3. Is there any link between the number of full-time employess and median salary?

By looking the 'Full Time' x 'Median Salary' plot, it's possible to conclude that there is no correlation between the number of full-time employess and median salary.

### Analysing the histograms

1. What's the most common median salary range?

To answer this question, we will observe the histogram about the median salaries. By looking it, we see that the salary range from 30000 to 40000 dollars is the most common in the data because this range has the highest bar.

2. What percent of majors are predominantly male? Predominantly female?

To answer these questions, we will examine the histogram about Sharewomen. The majors with Sharewomen < 0.5 represent the majors that are predominantly male. On the other side, the majors with Sharewomen >= 0.5 represent the majors that are predominantly female. We divided the data in ten bars, so the first 5 fives bars represent the male predominant majors, and the last five, the women predominant majors. Then, the heights of the bars are approximately [3, 14, 16, 22, 19, 21, 25, 29, 11, 12]. Then, the percentage of majors that are predominantly male is 43% and the percentage of the majors that are predominantly female is 57%.

## Pandas, Bar Plots
Let's explore the data with help of bar plots.

### Comparing the percentage of women from the 10 highest majors and from the 10 lowest paying majors

```python
# The majors are ordered by median salary, from highest to lowest

recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[len(recent_grads) - 10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

### Findings 
We can see that the most paying majors are prodominantly male. But, the lowest paying majors are predominantly female.

## Comparing the unemployment rate from the 10 highest paying majors and from the 10 lowest paying majors

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[len(recent_grads) - 10:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

### Findings
We can see that them unemployment rate of the lowest paying majors are a little bit higher than the unemployment rate of the highest paying majors.
